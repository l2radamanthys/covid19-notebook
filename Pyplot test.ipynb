```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 2
```

```python
x = np.linspace(-1, 5, num=30)
```

```python
fig, axes = plt.subplots()
axes.plot(x, f(x), 'ro', label='Funcion f(x)')
axes.set_xlim(-2, 6)
axes.set_ylim(-10, 30)
```

```python
fig.savefig('grafica1.png')
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
axes[0].plot(x, f(x), label='Funcion f(x)')
axes[1].plot(x, -f(x), 'red', label='Funcion f(x)')
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
s = 200 * np.random.randn(100)
c = np.random.randn(100)

plt.scatter(x, y, s, c, cmap=plt.cm.Blues_r)
```

```python
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
xx, yy = np.meshgrid(x, y)
```

```python
def g(x,y):
    return np.cos(x) + np.sin(y) ** 2
```

```python
zz = g(xx, yy)
```

```python
fig, axes = plt.subplots()
axes.contourf(xx, yy, zz, cmap=plt.cm.autumn)
```
